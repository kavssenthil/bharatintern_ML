```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
df = pd.read_csv("train.csv")
df
```

```python
indexs_null_cols = [index for (index,value) in enumerate(df.isnull().sum().values) if value>0]
```

```python
df.isnull().sum().iloc[indexs_null_cols]
```

```python
indexs = df.iloc[:,indexs_null_cols].dtypes.index
indexs
```

```python
for i in df.columns:
    if df[i].dtype == "object":
        df[i].replace(df[i].unique(),list(range(len(df[i].unique()))),inplace = True)
        df[i].astype(int)
    elif df[i].dtype == 'float64':
        df[i].fillna(0,inplace = True)
        
    
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
x.shape,y.shape
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3)
```

```python
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
pred = model.predict(x_test)
pred
```
